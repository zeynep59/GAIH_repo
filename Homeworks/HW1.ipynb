Differences of unsupervised and supervised learning:
    -In supervised learning given data to progam but in unsupervised learning does not.
    -The goal of supervised learning is to train the model so that it can predict the output when it is given new data.
The goal of unsupervised learning is to find the hidden patterns and useful insights from the unknown dataset.
    -Supervised learning model produces an accurate result but 
unsupervised learning model may produces less accurate result as compared to supervised learning.

ALgorithms for supervised learning :
    Linear regression
    Decision trees
    Logistic regression
Algorithms for unsupervised learning:
    K-means algorithm
    Hierarchical clustering algorithm
    Density based clustering algorithm

preprocessing steps:
    Importing necessary libraries
    Importing the datasets
    
    Handling null values: No model can handle any NULL or NaN values on its own so we need to intervene. 
we can check the null values with isnull() method  and we can delete the null values with .dropna() method.

    Standardization: transforming  values such that the mean of the values is 0 and the standard deviation is 1.
We can use the StandartScaler function for standardization.

    Handling categorical variables: Categorical variables are basically the variables that are discrete and not continuous.
Categorical variables are further divided into 2 types as nominal and ordinal variables. These variables need to preprocess differently. 
we can use map() or LabelEncoder() functions to handle ordinal categorical variables. Also we can use get_dummies()(from pandas library) function 
to handle nominal categorival variables.

    One-hot encoding: We create ’n’ columns where n is the number of unique values that the nominal variable can take in one-hot encoding.

    Multicollinearity: If we have multicollinearity in our dataset then we won’t be able to use our weight vector to calculate the feature importance. Multicollinearity impacts the interpretability of our model.
    
    The quality of data and the useful information that can be derived from it directly affects
the ability of our model to learn; therefore, it is extremely important that we preprocess our data before feeding it into our model.
